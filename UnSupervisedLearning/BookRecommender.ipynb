```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv("/kaggle/input/goodreadsbooks/books.csv",on_bad_lines='skip')
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df=df.drop(columns=['isbn','isbn13','publication_date'])
```

```python
df['content']=df['title']+' '+df['authors']+' '+df['publisher']
```

```python
tfidf=TfidfVectorizer(max_features=5000,stop_words='english')
vectors=tfidf.fit_transform(df['content'])
```

```python
similarities=cosine_similarity(vectors)
```

```python
def recommend(book,n_movies=15):
    book_indicies=df[df['title'].str.contains(book)].index
    if len(book_indicies)==0:
        print("Book not found in the dataset")
        return 
    idx=book_indicies[0]
    distances=similarities[idx]
    book_list=sorted(list(enumerate(distances)),key=(lambda x:x[1]),reverse=True)[1:n_movies]

    for i in book_list:
        print(df.iloc[i[0]].title)
    
```

```python
recommend('Harry')
```
